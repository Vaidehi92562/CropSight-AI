```python
import pandas as pd

X_final = pd.read_csv("../data/processed/X_final.csv")

final_features = X_final.columns.tolist()

final_features
```

```python
import joblib
model = joblib.load("risk_model.pkl")
```

```python
len(final_features)
```

```python
import numpy as np

# create template with zeros
template = np.zeros(len(final_features))

template
```

```python
scenarios = pd.DataFrame([
    template.copy(),
    template.copy(),
    template.copy()
], columns=final_features)

scenarios.index = ["Ideal", "Moderate", "Extreme"]
```

```python
print(final_features)
```

```python
def assign_risk_zone(score):
    if score <= 30:
        return "Low"
    elif score <= 60:
        return "Moderate"
    elif score <= 80:
        return "High"
    else:
        return "Critical"

scenarios["Risk Score"] = model.predict_proba(scenarios)[:,1] * 100
scenarios["Risk Zone"] = scenarios["Risk Score"].apply(assign_risk_zone)

scenarios
```

```python
def mitigation_recommendation(row):
    if row["Risk Score"] > 80:
        return "Immediate cold storage & emergency dispatch"
    elif row["Risk Score"] > 60:
        return "Priority routing and temperature-controlled storage"
    elif row["Risk Score"] > 30:
        return "Monitor logistics and optimize transport time"
    else:
        return "Standard handling procedures"

scenarios["Recommended Action"] = scenarios.apply(mitigation_recommendation, axis=1)

scenarios[["Risk Score","Risk Zone","Recommended Action"]]
```
